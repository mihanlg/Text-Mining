```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.cluster.hierarchy import linkage, fcluster
```

```python
points = np.array([[0.6, 1.9],
                   [1.8, 1.6],
                   [2.7, 2.0],
                   [3.0, 2.1],
                   [3.0, 2.6],
                   [3.1, 4.5],
                   [3.8, 0.6],
                   [4.2, 2.7]])
names = 'abcdefgh'
```

```python
def plot_cluster(colors = None, title = ''):
    global names, points
    plt.figure(figsize=(10,5))
    plt.title(title)
    plt.scatter(points[:,0], points[:,1], 60, c = colors)
    #plt.axis('off')
    for i in xrange(len(names)):
        plt.text(points[i][0]+0.05, points[i][1]+0.05, names[i])
```

# Оценим расположение точек

```python
plot_cluster()
```

# Single Linkage

```python
single = linkage(points, method='single', metric='cosine')
```

```python
for i in xrange(1, len(points)+1):
    clusters = fcluster(single, i, criterion='maxclust')
    plot_cluster(clusters, 'Single linkage, {0} cluster{1}, real number: {2}'.format(i, 's' if i > 1 else '',
                                                                                    len(np.unique(clusters))))
```

# Complete Link

```python
complete = linkage(points, method='complete', metric='cosine')
```

```python
for i in xrange(1, len(points)+1):
    clusters = fcluster(complete, i, criterion='maxclust')
    plot_cluster(clusters, 'Complete link, {0} cluster{1}, real number: {2}'.format(i, 's' if i > 1 else '',
                                                                                    len(np.unique(clusters))))
```
